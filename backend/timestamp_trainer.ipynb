```python
import pandas as pd

def get_most_active_hour(df):
    
    # Extract the hour of the day 
    df['hour'] = df['timestamp'].dt.hour

    # Grouping by customer ID and hour 
    df_grouped = df.groupby(['customer_id', 'hour']).size().unstack().fillna(0)

    # total number of actions performed
    df_grouped['total'] = df_grouped.sum(axis=1)

    # Sorting data by no. of actions
    df_grouped = df_grouped.sort_values(by=['total'], ascending=False)

    df_grouped['most_active_hour'] = df_grouped.iloc[:, :-1].idxmax(axis=1)
    
    df_grouped.reset_index(inplace=True)

    return df_grouped[['customer_id', 'most_active_hour']]

```

```python
df = pd.read_csv('datasets/embedded_dataset.csv')
pd.set_option('display.max_rows', None)
df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df_grouped = get_most_active_hour(df.copy())
df_grouped = df_grouped.sort_values(by=['customer_id']).reset_index(drop=True)
df_grouped.to_csv('most_active_hour_results.csv', index=False)
df_grouped.head(15)
```
